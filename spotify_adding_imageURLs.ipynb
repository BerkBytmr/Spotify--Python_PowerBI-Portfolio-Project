```python
import requests
import pandas as pd
from requests.auth import HTTPBasicAuth

# -------- AUTH --------
def get_spotify_token(client_id, client_secret):
    url = "https://accounts.spotify.com/api/token"
    response = requests.post(
        url,
        data={"grant_type": "client_credentials"},
        auth=HTTPBasicAuth(client_id, client_secret)
    )
    data = response.json()
    return data["access_token"]

# -------- SEARCH --------
def search_track(track_name, artist_name, token):
    query = f"{track_name} artist:{artist_name}"
    url = f"https://api.spotify.com/v1/search?q={query}&type=track&limit=1"

    response = requests.get(
        url,
        headers={"Authorization": f"Bearer {token}"}
    )

    data = response.json()

    try:
        return data["tracks"]["items"][0]["id"]
    except (KeyError, IndexError):
        return None

# -------- TRACK DETAILS --------
def get_track_details(track_id, token):
    url = f"https://api.spotify.com/v1/tracks/{track_id}"

    response = requests.get(
        url,
        headers={"Authorization": f"Bearer {token}"}
    )

    data = response.json()
    return data["album"]["images"][0]["url"]

# -------- MAIN --------
client_id = "9dad4165c2fc43bf817dc78f1123ed8e"
client_secret = "6eb9a1e93ee84982a4137b29f00e15f0"

access_token = get_spotify_token(client_id,client_secret )

df_spotify = pd.read_csv(
    r"C:\Users\HUAWEI\Desktop\spotify\spotify-2023.csv",
    encoding="ISO-8859-1"
)

df_spotify["image_url"] = None

for i, row in df_spotify.iterrows():
    track_id = search_track(
        row["track_name"],
        row["artist(s)_name"],
        access_token
    )
    if track_id:
        df_spotify.at[i, "image_url"] = get_track_details(track_id, access_token)

df_spotify.to_csv('updated_file.csv', index=False)
```
